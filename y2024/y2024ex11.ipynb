```python
#@title part 1
from numba import jit
from collections import defaultdict

@jit
def figures(n: int) -> int:
    ret=0
    if n==0:
        return 1
    while n>0:
        ret+=1
        n//=10
    return ret


def step(data):
    ret=defaultdict(int)
    for n, times in data.items():
        if n==0:
            ret[1]+=times
            continue

        f = figures(n)
        if f&1:
            ret[n*2024]+=times

        else:
            m =10**(f//2)
            ret[n//m]+=times
            ret[n%m]+=times

    return ret

def part1(data):
    for i in range (25):
        data=step(data)
    return sum(data.values())

print(part1(parser(readme_input)))
print(part1(parser(_input)))
```

```python
#@title part 2
def part2(data):
    for i in range (75):
        data=step(data)
    return sum(data.values())

print(part2(parser(readme_input)))
print(part2(parser(_input)))
```
